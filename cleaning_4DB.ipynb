```python
df=df.drop(columns=['TMC','End_Lat','End_Lng','Distance(mi)',
       'Description','Number', 'Street','Zipcode','Country', 'Timezone', 'Airport_Code', 'Weather_Timestamp',
       'Temperature(F)','Wind_Chill(F)', 'Humidity(%)', 'Pressure(in)','Wind_Direction','Precipitation(in)', 'Amenity',
                                            'Civil_Twilight', 'Nautical_Twilight',
       'Astronomical_Twilight'])
```

```python
df_severity=pd.DataFrame(
    data=df,columns=['ID','Severity'],index=df.index.copy())
df_severity
```

```python
df_accidenttime=pd.DataFrame(
    data=df,columns=['ID','Start_Time', 'End_Time','Sunrise_Sunset'],index=df.index.copy())
df_accidenttime
```

```python
df_accidentloc=pd.DataFrame(
    data=df,columns=['ID','Start_Lat',
       'Start_Lng', 'Side', 'City', 'County'],index=df.index.copy())
df_accidentloc
```

```python
df_accidentweather=pd.DataFrame(
    data=df,columns=['ID','Visibility(mi)',
       'Wind_Speed(mph)', 'Weather_Condition'],index=df.index.copy())
df_accidentweather
```
